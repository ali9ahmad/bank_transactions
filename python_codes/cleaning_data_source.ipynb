First we install needed libraries.

Then, we upload needed json file

```python
import pandas as pd

df2 = pd.read_json("transactions_3000.json")

df2
```

After importing data and making sure everything is working.
We need to work with data frame. Doing the works like: Cleaning, Sorting. 

1. We need to check if we have any null or nan values. if we have them we need to delete those. Since they do not provide any valuable info.

```python
df2.isnull().sum()
df2.isna().sum()
```

After that. We check for all of the negative amount value in amount column.

```python
# df[df["amount"] <= 0]

df2 = df2[df2["amount"] > 0]
```

After removing the negative value we need to check the date of the transactions since we are trying to check only the current year we will remove the transactions happened before 2024.

```python
df2 = df2[df2["date"].dt.year == 2025]

df2
```

Since now we have clean data we can export it to csv file to upload it to our sql database afterwords.

```python
df2.to_csv("cleaned_transactions.csv", index=False)
```
