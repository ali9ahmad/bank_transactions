first we read the file


```python
import pandas as pd

df1 = pd.read_csv("branch_logs.txt", sep=r"\s+-\s+",   header=None, engine="python")

df1
```

when we got it to look of dataframe we start working with the file. we can start by giving column names


```python
df1.columns = ["account_id", "customer_id", "branch_name", "date", "transaction_type", "amount"]

df1
```

now we need to remove not needed words from the rows. words like account and branch

```python
df1["account_id"] = df1["account_id"].str.extract(r'Account (\d+)')

df1
```

```python
df1["branch_name"] = df1["branch_name"].str.extract(r'Branch:\s*(.+)')

df1
```

```python
df1["amount"] = df1["amount"].str.replace("$", "", regex=False).astype(float)

df1
```

```python
df1.isnull().sum()
df1.isna().sum()
```

after completing working with format now we check for negative values and info for only happened in 2025

```python
df1["date"] = pd.to_datetime(df1["date"], format="%Y-%m-%d", errors="coerce")
df1.info()
```

```python
df_2024 = df1[df1["date"].dt.year == 2024]

df_2024

# df["date"].dt.year.value_counts()

df1 = df1[df1["date"].dt.year == 2025]

df1
```

since we cleaned the data from prvs year now we need to remove negative data

```python
# df[df["amount"] <= 0].count()

df1 = df1[df1["amount"] > 0]

df1
```

since now we have only positive data and period for 2025 we can export this data to csv to later upload to our database


```python
df1.to_csv("branch_logs.csv", index=False)
```
