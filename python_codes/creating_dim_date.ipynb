```python
import pandas as pd 

dates = pd.date_range(start = "2025-01-01", end = "2025-12-31")

dim_date = pd.DataFrame({
    'date': dates,
    'year': dates.year,
    'month': dates.month,
    'day': dates.day,
    'quarter': dates.quarter,
    'day_name': dates.day_name(),
    'month_name': dates.month_name(),
}
)

# dim_date.set_index('date', inplace=True)
dim_date = dim_date.reset_index()
dim_date.insert(0,'date_key',range(1, len(dim_date) + 1))
dim_date = dim_date.drop('index', axis=1)
dim_date
```

as of now we have good dim_date table we can export it to csv file to later use it in our ssms



```python
dim_date.to_csv("dim_date.csv", index=False)
```
